```python
#import libraries
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt



```

```python
#import data 
data = pd.read_csv('./data/data.csv')
data.head()
# data.tail()
# data.columns
# data.shape
```

```python
#creating copy of data for safe practice 
data_copy = data.copy()
data_copy
```

```python
#dropping column Road Type
data_copy.drop(columns=['Road'], inplace=True)

data_copy.head()
```

```python
#removing the first word in the "Address" column and set only the remaining place name
data_copy['Address'] = data_copy['Address'].apply(lambda x: ', '.join(x.split(', ')[-2:]))
data_copy.head(50)
```

```python
#removing decimal from year column
data_copy['Year'] = data_copy['Year'].fillna(0).astype(int)
data_copy.head()
```

```python
#dropping column Road Type
data_copy.drop(columns=['City'], inplace=True)

data_copy.head()
```

```python
# Replace 'N/A' with '0' in 'Build Area' column
data_copy['Build Area'] = data_copy['Build Area'].str.replace('N/A', '0', regex=False)

# Display the first 50 rows
data_copy.head(50)
```

```python
data_copy['Floors'] = data_copy['Floors'].fillna(0)

data_copy.head(50)
```

```python
# Function to convert various formats to square feet
def convert_to_square_feet(area):
    try:
        if 'Aana' in area:
            parts = area.replace(' Aana', '').split('-')
            # Ensure all parts are present
            parts = [int(p) if p else 0 for p in parts] + [0] * (4 - len(parts))
            ropani, aana, paisa, dam = parts
        elif 'Ropani' in area:
            # Split the string into parts
            parts = area.replace(' Ropani', '').replace(' Aana', '').replace(' Paisa', '').replace(' Dam', '').split('-')
            ropani, aana, paisa, dam = (int(parts[i]) if i < len(parts) else 0 for i in range(4))
        elif 'Sq. Feet' in area:
            return float(area.replace(' Sq. Feet', ''))
        else:
            return 0  # Handle 'N/A' or other unexpected cases
        
        # Conversion factors
        ropani_to_sqft = 5476
        aana_to_sqft = 342.25
        paisa_to_sqft = 85.56
        dam_to_sqft = 21.39
        
        # Calculate total square feet
        total_sqft = (ropani * ropani_to_sqft) + (aana * aana_to_sqft) + (paisa * paisa_to_sqft) + (dam * dam_to_sqft)
        return total_sqft
    except Exception as e:
        return None

# Apply the conversion function to 'Build Area' and 'Area' columns
data_copy['Build Area'] = data_copy['Build Area'].apply(convert_to_square_feet)
data_copy['Build Area'] = data_copy['Build Area'].fillna(0)

data_copy['Area'] = data_copy['Area'].apply(convert_to_square_feet)
data_copy['Area'] = data_copy['Area'].fillna(0)

# Display the updated DataFrame
data_copy.head(40)
```

```python
data_copy['Road Type'] = data_copy['Road Type'].where(pd.notnull(data_copy['Road Type']), None)

# Display the updated DataFrame
data_copy.head(50)
```

```python
# Function to convert road width to feet
def convert_to_feet(width):
    try:
        if isinstance(width, str):
            if 'Feet' in width:
                return float(width.replace(' Feet', ''))
            elif 'Meters' in width:
                return float(width.replace(' Meters', '')) * 3.28084  # Convert meters to feet
        return float(width)
    except Exception as e:
        return None

# Apply the conversion function to 'Road Width' column
data_copy['Road Width'] = data_copy['Road Width'].apply(convert_to_feet)

# Display the updated DataFrame
data_copy.head(20)
```

```python
data_copy.to_csv('data_copy.csv', index=False)    
```
