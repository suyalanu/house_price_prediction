```python
#import libraries
import pandas as pd
import sklearn 
```

```python
#import data 
data = pd.read_csv('./data/data_copy.csv')
data.head()
```

```python
data_copy = data.copy()
data_copy
```

```python
data_copy['Build Area'] = data_copy['Build Area'].fillna('0-0-0-0 Aana')
data_copy['Build Area'].head()
```

```python
# Function to convert various formats to square feet
def convert_to_square_feet(area):
    try:
        if 'Aana' in area:
            parts = area.replace(' Aana', '').split('-')
            # Ensure all parts are present
            parts = [int(p) if p else 0 for p in parts] + [0] * (4 - len(parts))
            ropani, aana, paisa, dam = parts
        elif 'Ropani' in area:
            # Split the string into parts
            parts = area.replace(' Ropani', '').replace(' Aana', '').replace(' Paisa', '').replace(' Dam', '').split('-')
            ropani, aana, paisa, dam = (int(parts[i]) if i < len(parts) else 0 for i in range(4))
        elif 'Sq. Feet' in area:
            return float(area.replace(' Sq. Feet', ''))
        else:
            return 0  # Handle 'N/A' or other unexpected cases
        
        # Conversion factors
        ropani_to_sqft = 5476
        aana_to_sqft = 342.25
        paisa_to_sqft = 85.56
        dam_to_sqft = 21.39
        
        # Calculate total square feet
        total_sqft = (ropani * ropani_to_sqft) + (aana * aana_to_sqft) + (paisa * paisa_to_sqft) + (dam * dam_to_sqft)
        return total_sqft
    except Exception as e:
        return None

# Apply the conversion function to 'Build Area' and 'Area' columns
data_copy['Build Area Sq. Feet'] = data_copy['Build Area'].apply(convert_to_square_feet)
data_copy['Build Area Sq. Feet'] = data_copy['Build Area Sq. Feet'].fillna(0)

data_copy['Area Sq. Feet'] = data_copy['Area'].apply(convert_to_square_feet)
data_copy['Area Sq. Feet'] = data_copy['Area Sq. Feet'].fillna(0)

# Display the updated DataFrame
data_copy.head(40)
```

```python
data_copy.drop(columns=['Area','Build Area'], inplace=True)

data_copy.head(50)
```

```python
data_copy.to_csv('data_copy.csv', index=False)    
```
